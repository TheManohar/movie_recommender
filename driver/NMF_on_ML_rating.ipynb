```python
long = df.groupby(['userId', 'movieId'])['rating'].mean()
long.head()
```

```python
wide = long.unstack()
```

```python
wide = wide.fillna(0.0)
wide.info()
```

```python
R = wide
R
```

```python
# model assumes R ~ PQ'
model = NMF(n_components=2, init='random')
model.fit(R)
```

```python
P = model.components_  # movie feature
P
```

```python
Q = model.transform(R)  # user feature
Q
```

```python
# Value of the Loss Fuction
model.reconstruction_err_
```
