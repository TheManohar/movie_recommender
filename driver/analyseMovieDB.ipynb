```python
df = qGroupLens("SELECT movieId, imdbId, userId, title, genres, rating FROM moviesGroupLens WHERE userID !='NaN' ORDER BY userId ASC")
#dbGroupLens("SELECT * FROM moviesGroupLens WHERE isOriginalTitle=1 AND averageRating>9 AND numVotes>40000 GROUP BY id ORDER BY averageRating DESC")
#movieTitles = df['title']
#userId = df['userId']
userMovieRating = df[['movieId', 'userId', 'rating']]
#print(userMovieRating.head(20))

userMovieRating.set_index(['movieId','userId'], inplace=True)
userMovieRating = userMovieRating.stack(0).unstack(1)
#userMovieRating = userMovieRating.unstack(1)
userMovieRating.fillna(0, inplace=True)
print(userMovieRating.head(5))

movies = userMovieRating.columns
print(movies)
users = userMovieRating.index
print(users)
```

```python
userR = userMovieRating.values
```

```python
# model assumes R ~ PQ'
model = NMF(n_components=2, init='random', random_state=10)
print(model.fit(userR))
```

```python
P = model.components_  # movie feature
print(P)
```

```python
Q = model.transform(userR)  # user feature
print(Q)
```

```python
print(model.reconstruction_err_)
```

```python
nR = np.dot(Q, P)
print(nR.T)
#movies = ['Titanic', 'Tiffany', 'Terminator', 'Star Trek', 'Star Wars']
#users = ['u1', 'u2', 'u3', 'u4']
nRDf = pd.DataFrame(nR.T)#, columns = movies, index = users)

import seaborn as sns
sns.heatmap(nRDf, annot=True)
plt.show()
```
