```python
DB_SETUP = """
    CREATE TABLE IF NOT EXISTS lite_moviesDB (
        id VARCHAR(255),
        ordering FLOAT,
        title VARCHAR(255),
        language VARCHAR(255),
        isOriginalTitle BOOLEAN,
        averageRating FLOAT,
        numVotes INTEGER,
        titleType VARCHAR(255),
        primaryTitle VARCHAR(255),
        originalTitle VARCHAR(255),
        startYear INTEGER,
        runtimeMinutes INTEGER,
        genres VARCHAR(255)
        );"""
```

```python
db = sqlite3.connect('lite_movies.db')
db.executescript(DB_SETUP)
```

```python
for i, row in df3.iterrows():
        query = 'INSERT INTO lite_moviesDB VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)'
        db.execute(query, (
            i, 
            row['ordering'], 
            row["title"], 
            row["language"], 
            row["isOriginalTitle"], 
            row["averageRating"], 
            row["numVotes"], 
            row["type"],
            row["primaryTitle"],
            row["originalTitle"],
            row["year"],
            run["runtimeMinutes"],
            row["genres"]
        ))

db.commit()
db.close()

print('\n### Congrats! Database movies.db created! ###')
```
